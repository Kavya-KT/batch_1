## merge()

```python

df1 = pd.DataFrame({
    'empid': [101, 102, 103, 104, 105],
    'empname': ['Anand', 'bavish', 'catherine', 'davis', 'elias'],
    'designation': ['doctor', 'engineer', 'advocate', 'teacher', 'developer']})
```

```python
df1
```

```python
df2 = pd.DataFrame({
    'empid': [101, 102, 103, 106, 107],
    'salary': [20000, 20300, 30000, 23000, 27000],
    'location': ['ern', 'kollam', 'thrissur', 'aluva', 'cherthala']})
df2
```

## inner merge()

```python
df=pd.merge(df1,df2,how='inner',on='empid')
df
```

## outer merge

```python
-->Keeps all rows from both DataFrames.
-->Fills NaN for missing matches on either side.
```

```python
df=pd.merge(df1,df2,how='outer',on='empid')
df
```

## right merge

```python
df=pd.merge(df1,df2,how='right',on='empid')
df
```

## left merge

```python
df=pd.merge(df1,df2,how='left',on='empid')
df
```
