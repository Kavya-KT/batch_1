```python
#1.Give column name as Id, Movie, Year, Rating, Duration.
df=pd.read_csv(r"D:\important\dataspark\data science\7.Machine learning\Datasets\movies_cleaned.csv",names=['Id', 'Movie', 'Year', 'Rating', 'Duration'])
df
```

```python
#2.Display the number of rows and number of columns.
df.info()
```

```python
df1=df.copy()
```

```python
#3.Display the five observations that have highest duration.
df1.sort_values(by='Duration').tail()
```

```python
#4.Remove if any duplicate exists and change the index accordingly.
df2=df1.copy()
df2.drop('Id',axis=1,inplace=True)
df2.drop_duplicates(inplace=True)
 
```

```python
df2.reset_index(drop=True,inplace=True)
df
```

```python
#5.Rename the column name ‘Movie’ as Film.

#dfname.rename({old:new},axis=1)
df2.rename({'Movie':'Film'},axis=1,inplace=True)
df2
```

```python
#6.Display all the observation starting from index 100 upto 200.
df2.iloc[100:201]
```

```python
 #7.Display the name and releasing year of the film with maximum duration.
df2.sort_values(by='Duration').tail(1)[['Film','Year']]
```

```python
#8.Display the count of films released in each year.
df2[['Year']].value_counts()
```

```python
#9.Find the film released between 1986–1996 having highest duration.
df2.loc[(df2['Year']>=1986)&(df2['Year']<=1996)].sort_values(by='Duration').tail(1)[['Film']]
```

```python
#10Find the films having duration greater than 7000 and rating above 3.5.
df2.loc[(df2['Duration']>=7000)&(df2['Rating']>=3.5)]
```

```python
#11.Print the details about 5 films having less rating, display in the descending order.
df2.sort_values(by='Rating',ascending=False).tail()
```
